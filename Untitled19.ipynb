$$\Large\color{red}{\text{Caminata Aleatoria en 1D}}$$

Una caminata aleatoria en una dimensión consiste en una partícula que se mueve en pasos discretos sobre la recta real.

En cada paso:

- Se mueve +1 a la derecha con probabilidad \( p \)
- Se mueve −1 a la izquierda con probabilidad \( q \)

cumpliendo que:

$
p + q = 1
$

Si definimos la posición después de \( n \) pasos como:

$$
X_n = \sum_{i=1}^{n} S_i
$$

donde cada $ S_i \in \{-1, +1\} $, entonces:

- Si \( p = 0.5 \) → Caminata simétrica  
- Si \( p $\neq$ 0.5 \) → Caminata con sesgo

---

## Tiempo de Primer Retorno

El tiempo de primer retorno al origen se define como:

$
T_0 = \min \{ n \geq 1 : X_n = 0 \}
$

Para caminata simétrica:

$
P(T_0 < \infty) = 1
$

Es decir, la partícula regresa al origen con probabilidad 1.


## Explicación del código

El código simula una caminata aleatoria en una dimensión. Primero se importan las librerías necesarias: $\texttt{numpy}$ para generar los pasos aleatorios y hacer operaciones con arreglos, y $\texttt{matplotlib}$ para poder graficar la trayectoria.

Después se definen los parámetros principales: \(n\) es el número de pasos que dará la partícula y \(p\) es la probabilidad de moverse a la derecha \((+1)\), mientras que \(1-p\) es la probabilidad de moverse a la izquierda \((-1)\).

La función \(\texttt{caminata\_aleatoria(n,p)}\) genera los pasos usando una selección aleatoria de valores \(+1\) y \(-1\) con las probabilidades dadas. Luego, con la suma acumulada

$
X_n = \sum_{i=1}^{n} S_i
$

se obtiene la posición en cada paso. También se agrega el valor inicial $X_0 = 0$ para que la caminata comience desde el origen.

Finalmente, el programa imprime los resultados y grafica la trayectoria, donde se puede observar cómo la partícula se mueve a lo largo del tiempo dependiendo de los pasos generados.


```python
import numpy as np
import matplotlib.pyplot as plt
n = 200 #pasos
p = 0.5  #
q = 1 - p # formula
num_tray = 1  #trayectoria


def caminata_aleatoria(n, p):

    pasos = np.random.choice([1, -1], size=n, p=[p, 1-p])
    posicion = np.cumsum(pasos)

    posicion = np.insert(posicion, 0, 0)

    return pasos, posicion

pasos, trayectoria = caminata_aleatoria(n, p)

print("Pasos generados:\n", pasos)

```

```python
plt.figure(figsize=(10,6))

for i in range(num_tray):
    tray = caminata_aleatoria(n, p)
    plt.plot(tray)

plt.title("Trayectorias de Caminata Aleatoria 1D")
plt.xlabel("Pasos")
plt.ylabel("Posición")
plt.grid(True)

plt.show()
```

# Análisis y Conclusiones

A partir de las simulaciones realizadas se observa que:

- Cuando \( p = 0.5 \), la caminata es simétrica.
- La partícula oscila alrededor del origen.
- Eventualmente regresa al punto inicial.

Cuando \( p > 0.5 \):

- Existe un sesgo hacia la derecha.
- Las trayectorias tienden a crecer.
- La probabilidad de retorno disminuye.

Cuando \( p < 0.5 \):

- El sesgo es hacia la izquierda.
- La partícula se aleja en dirección negativa.

La caminata aleatoria 1D permite modelar fenómenos como:

El parámetro \( p \) controla el comportamiento global del sistema, determinando si el proceso es recurrente o transitorio.

## 7¿Cuál es la probabilidad de que después de \( n \) pasos esté en una posición positiva?

Sea $ X_n $ la posición después de \( n \) pasos:

$
X_n = \sum_{i=1}^{n} S_i,
\quad
S_i \in \{-1,+1\}
$

Entonces:

$
X_n = 2K - n,
\quad
K \sim Binomial(n,p)
$

La probabilidad buscada es:

$
P(X_n > 0)
=
\sum_{k=\lfloor n/2 \rfloor + 1}^{n}
\binom{n}{k}
p^k (1-p)^{n-k}
$

Si \( p = 0.5 \):

$
P(X_n > 0) \approx 0.5
$

---

## ¿Cuál es la probabilidad de que la posición máxima supere cierto valor \( a \)?

Sea:

$
M_n = \max_{1 \le k \le n} X_k
$

Se desea:

$
P(M_n \ge a)
$

Para caminata simétrica \( (p=0.5) \):

$
P(M_n \ge a) = 2\,P(X_n \ge a)
$

Usando aproximación normal:

$
P(M_n \ge a)
\approx
2\left[
1 -
\Phi\!\left(\frac{a}{\sqrt{n}}\right)
\right]
$

donde $ \Phi $ es la CDF normal estándar.
